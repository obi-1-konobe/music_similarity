```python
with open('../data/dict_errors.txt', 'r') as f:
    err_list = f.readlines()
err_list = [int(x.strip('\n')) for x in err_list]
```

```python
with open('../data/pickles/main_dict.pickle', 'rb') as f:
    main_dict = pickle.load(f)
```

```python
main_dict[111399]
```

```python
i = 0
for err in err_list:
    try:
        artist = network.get_artist(main_dict[err]['artist_name'])
        artist_tags = artist.get_top_tags(10)
        main_dict[err]['artist_tags'] = [(elem.item.name, elem.weight) for elem in artist_tags]
        print('add artist tags')
    except Exception as e:
        print(e)
        continue
    try:    
        track = network.get_track(artist, main_dict[err]['track_title'])
        track_tags = track.get_top_tags(10)
        similar_track = track.get_similar()
    except Exception as e:
        print(e)
        continue
    else:
        
        main_dict[err]['track_tags'] = [(elem.item.name, elem.weight) for elem in track_tags]
        main_dict[err]['similar_tracks'] = [(elem.item.artist.name, elem.item.title, elem.match) for elem in similar_track]
        i += 1
i
    
```

```python
with open('../data/pickles/main_dict.pickle', 'wb') as f:
    pickle.dump(main_dict, f)
```
