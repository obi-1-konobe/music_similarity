```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('../data/csv/track_vectors_1.csv', index_col=0)
```

```python
features = df.drop(['track'], 1)
```

```python
 features
```

```python
st_feat = StandardScaler().fit_transform(features)
```

```python
st_feat
```

```python
pca_viz = PCA(n_components=2)
```

```python
two_pca = pca_viz.fit_transform(st_feat)
```

```python
two_pca
```

```python
principalDf = pd.DataFrame(data = two_pca
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
pca_viz.explained_variance_ratio_
```

```python
fig = plt.figure(figsize = (8,8))

ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)

ax.scatter(two_pca.T[0], two_pca.T[1])
ax.grid()
```

```python
pca_viz_3 = PCA(n_components=3)
```

```python
three_pca = pca_viz_3.fit_transform(st_feat)
```

```python
three_pca
```

```python
pca_viz_3.explained_variance_ratio_
```

```python
fig = plt.figure(1, figsize=(10, 10))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(three_pca[:, 0], three_pca[:, 1], three_pca[:, 2])
```
